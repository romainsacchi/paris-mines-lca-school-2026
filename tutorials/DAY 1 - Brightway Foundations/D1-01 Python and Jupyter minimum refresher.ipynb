## 1) Variables and lists

Lists are useful whenever we want to store several values of the same kind, such as impact scores or exchange amounts.


```python
impact_scores = [2.1, 3.4, 1.8, 2.9]
scaled_scores = [score * 1.1 for score in impact_scores]

print('Original scores:', impact_scores)
print('Scaled scores:', scaled_scores)
print('Average scaled score:', sum(scaled_scores) / len(scaled_scores))
```

## 2) Tuples

Tuples are useful for fixed ordered information, such as a Brightway method key or a database-plus-flow identifier. Like lists, tuples keep order, but they are typically used when the content should stay unchanged.


```python
method_key = ('IPCC 2021', 'climate change', 'GWP 100a')
biosphere_flow = ('biosphere3', 'Carbon dioxide, fossil')

print('Method key:', method_key)
print('Impact category:', method_key[1])

database, flow_name = biosphere_flow
print('Flow identifier:', database, '->', flow_name)
```

## 3) Dictionaries

Dictionaries are convenient when we want to store labeled information, for example the metadata of an activity or a set of characterization factors.
They contain key-value pairs, where each key is unique and maps to a specific value. 


```python
activity = {
    'name': 'Transport, passenger car, fleet average',
    'unit': 'person kilometer',
    'location': 'CH',
}

for key, value in activity.items():
    print(key, '->', value)
```

## 4) Loops and functions

Functions help us package a small reusable calculation.
We will use functions often later when we want to repeat the same LCA operation on several activities or scenarios.


```python
# First define the function with all arguments it should take. 
# In this case, we want to calculate the total impact based on flows and characterization factors (cfs).

def total_impact(flows, cfs):
    return sum(amount * cfs.get(flow, 0) for flow, amount in flows.items())

# Now we can use this function to calculate the total impact for a given set of flows and characterization factors.
# We first define the flows and characterization factors as dictionaries, where the keys are the flow names and the values are the amounts or factors.
flows = {'co2': 4.0, 'ch4': 0.1}
cfs = {'co2': 1.0, 'ch4': 27.0}

# Finally, we call the total_impact function with the defined flows and characterization factors and print the result.
print('Total impact:', total_impact(flows, cfs))
```

## 5) NumPy arrays

NumPy arrays are useful for numerical data when we want fast element-wise calculations and matrix-style operations. We will return to them later when we represent inventories and characterization factors as vectors and matrices.


```python
import numpy as np

# let's assume an intervention matrix for two activities
intervention = np.array([
    [4.0, 1.5],
    [0.1, 0.0],
    [0.02, 0.01],
])
cfs = np.array([1.0, 27.0, 273.0])

print('Intervention shape:', intervention.shape)
print('CF vector shape:', cfs.shape)
print('Scaled intervention:')
print(intervention * 2)
print('Characterized column totals:', cfs @ intervention) # @ is the matrix multiplication operator in Python, which multiplies the characterization factors with the intervention matrix to get the total impacts for each activity.
```
