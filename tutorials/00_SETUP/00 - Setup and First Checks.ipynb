## 4) Import checks in the current kernel

The Paris course uses one shared environment for `brightway`, `edges`, `premise`, `activity-browser`, and optional `trails` work.
The check below confirms that the main Python packages import cleanly in the current notebook kernel.


````python
import importlib

packages = [
    'bw2data',
    'bw2calc',
    'bw2io',
    'edges',
    'premise',
    'activity_browser',
    'dotenv',
]

for package in packages:
    try:
        module = importlib.import_module(package)
        print(f'{package}: OK -> {getattr(module, "__version__", "no __version__ attribute")}')
    except Exception as exc:
        print(f'{package}: FAILED -> {type(exc).__name__}: {exc}')
````

## Troubleshooting quick list

- If imports fail, first confirm that the notebook kernel is using the `lca-course` environment.
- If `premise` imports but later complains about missing credentials, confirm that the repo-root `.env` file exists and rerun the Day 3 or Day 4 `premise` notebook from the top so `load_dotenv()` is executed.
- If `bw2io.import_ecoinvent_release()` fails later in the course, verify that `.env` contains valid `ECOINVENT_USERNAME` and `ECOINVENT_PASSWORD` values, or enter them correctly when prompted.
- If `activity_browser` fails to import on Windows, recreate the environment from the YAML file instead of mixing manual `pip install` calls.
