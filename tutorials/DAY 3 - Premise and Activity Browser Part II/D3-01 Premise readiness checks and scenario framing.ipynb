## 3) Name the outputs before generating them

Keeping a predictable naming scheme makes the later comparisons in Brightway and Activity Browser much easier.


```python
def slugify(value):
    return value.lower().replace(' ', '-').replace('_', '-').replace('.', '-').replace('---', '-').replace('--', '-')

generated_db_names = [
    f"paris-{slugify(row.model)}-{slugify(row.pathway)}-{row.year}"
    for row in scenario_table.itertuples(index=False)
]
export_name = 'paris-trails-datapackage'

pd.DataFrame({'database_name': generated_db_names})
```
